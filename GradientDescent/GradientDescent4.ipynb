```python
import matplotlib.pyplot as plt

def neural_network(input, weights):
    out = 0
    for i in range(len(input)):
        out += (input[i] * weights[i])
    return out

def ele_mul(scalar, vector):
    out = [0,0,0]
    for i in range(len(out)):
        out[i] = vector[i] * scalar
    return out

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]
true = win_or_lose_binary[0]

alpha = 0.01
weights = [0.1, 0.2, -0.1]
input = [toes[0], wlrec[0], nfans[0]]

# Lists to hold error and weight values for plotting
error_list = []
weights_list = [[], [], []]

for iter in range(100):  # Increase the number of iterations for more data points
    pred = neural_network(input, weights)

    error = (pred - true) ** 2
    delta = pred - true

    weight_deltas = ele_mul(delta, input)

    # Append error and weight data to lists
    error_list.append(error)
    for i, w in enumerate(weights):
        weights_list[i].append(w)

    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]

# Plotting
plt.figure(figsize=(12, 6))

# Plot error
plt.subplot(1, 2, 1)
plt.title("Error")
plt.xlabel("Iteration")
plt.ylabel("Error")
plt.plot(error_list)

# Plot weights
plt.subplot(1, 2, 2)
plt.title("Weights")
plt.xlabel("Iteration")
plt.ylabel("Weight")
for i, weight_data in enumerate(weights_list):
    plt.plot(weight_data, label=f"Weight {i}")
plt.legend()

plt.tight_layout()
plt.show()
```
