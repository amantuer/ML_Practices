```python
import matplotlib.pyplot as plt

weights = [1.3, 0.2, 0.9]
weight_history = []
error_history = []

def neural_network(input, weights):
    return ele_mul(input, weights)

def ele_mul(scalar, vector):
    out = [0, 0, 0]
    for i in range(len(out)):
        out[i] = vector[i] * scalar
    return out

def scalar_ele_mul(number, vector):
    output = [0, 0, 0]
    for i in range(len(vector)):
        output[i] = number * vector[i]
    return output

wirec = [0.65, 1.0, 1.0, 0.9]
hurt = [0.1, 0.0, 0.0, 0.1]
win = [1, 1, 0, 1]
sad = [0.1, 0.0, 0.1, 0.2]

alpha = 0.1
epochs = 100

for epoch in range(epochs):
    for i in range(len(wirec)):
        input = wirec[i]
        true = [hurt[i], win[i], sad[i]]
        
        # Forward Pass
        pred = neural_network(input, weights)

        # Calculate Error and Delta
        error = [0, 0, 0]
        delta = [0, 0, 0]
        for j in range(len(true)):
            error[j] = (pred[j] - true[j]) ** 2
            delta[j] = pred[j] - true[j]
        
        # Record the average error for this iteration
        error_history.append(sum(error) / len(error))
        
        # Calculate weight deltas and update weights
        weight_deltas = scalar_ele_mul(input, delta)
        for j in range(len(weights)):
            weights[j] -= (weight_deltas[j] * alpha)
        
        # Record the updated weights
        weight_history.append(weights[:])

# Plotting the weights
plt.figure()
for i in range(len(weights)):
    plt.plot([w[i] for w in weight_history], label=f"Weight {i+1}")
plt.title("Weights Over Time")
plt.xlabel("Iteration")
plt.ylabel("Weight Value")
plt.legend()
plt.show()

# Plotting the error
plt.figure()
plt.plot(error_history, label="Error")
plt.title("Error Over Time")
plt.xlabel("Iteration")
plt.ylabel("Average Error")
plt.legend()
plt.show()
```
